```python
import os
import pandas as pd
import numpy as np
from datetime import datetime
import re


col = ['店铺名称'] + data.columns 
df=pd.DataFrame(columns= col )  #合起来的表

result_list = []
for parent,dirnames,filenames in os.walk(r'data/taobao/new/电器数据/'):
    for filename in filenames:
        #root 所指的是当前正在遍历的这个文件夹的本身的地址
        #dirs 是一个 list ，内容是该文件夹中所有的目录的名字(不包括子目录)
        #files 同样是 list , 内容是该文件夹中所有的文件(不包括子目录)
        result = re.match('^全店宝贝_(.*)_2021-05*',filename)
#         print(result.group(1))
        df1= pd.read_csv(parent+filename,skiprows =1) 
        result_list.append(result.group(1))
        df1['店铺名称'] =  result.group(1)
        df1 = df1.iloc[:len(df1)-4]
        df = pd.concat([df,df1])


df = df.set_index(['店铺名称','序号'])
df
df.to_csv(r'data/taobao/new/dianqi.csv',encoding="utf_8_sig")
```
